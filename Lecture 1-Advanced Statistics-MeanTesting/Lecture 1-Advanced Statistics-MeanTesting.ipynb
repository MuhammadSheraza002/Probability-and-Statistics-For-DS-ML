<h1 align='center'>Single Sample: Tests Concerning a Single Mean</h1> 

> **Example**
A random sample of 100 recorded deaths in
the United States during the past year showed an
average life span of 71 8 years Assuming a population
standard deviation of 8 9 years, does this seem to
indicate that the mean life span today is greater than
70 years? Use a 0 05 level of significance

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given data
sample_mean = 71.8
population_std_dev = 8.9
sample_size = 100
population_mean_hypothesized = 70
significance_level = 0.05

# Hypotheses
# Null hypothesis (H0): The mean life span today is less than or equal to 70 years.
# Alternative hypothesis (H1): The mean life span today is greater than 70 years.

# Calculate the test statistic
t_statistic = (sample_mean - population_mean_hypothesized) / (population_std_dev / np.sqrt(sample_size))

# Degrees of freedom
df = sample_size - 1

# Critical t-value for a one-tailed test
t_critical = stats.t.ppf(1 - significance_level, df)

# Print the hypotheses
print(f"Null Hypothesis (H0): μ ≤ {population_mean_hypothesized} (The mean life span today is less than or equal to {population_mean_hypothesized} years.)")
print(f"Alternative Hypothesis (H1): μ > {population_mean_hypothesized} (The mean life span today is greater than {population_mean_hypothesized} years.)\n")

# Print the results and decision
print(f"Sample Mean: {sample_mean}")
print(f"Population Standard Deviation: {population_std_dev}")
print(f"Sample Size: {sample_size}")
print(f"Population Mean (Hypothesized): {population_mean_hypothesized}")
print(f"Significance Level: {significance_level}")
print(f"Degrees of Freedom: {df}")
print(f"Critical t-value: {t_critical}")
print(f"Observed t-statistic: {t_statistic}")

# Decision based on the comparison
if np.abs(t_statistic) > t_critical:
    print(f"\nReject the null hypothesis. There is evidence to suggest that the mean life span today is greater than {population_mean_hypothesized} years.")
else:
    print(f"\nFail to reject the null hypothesis. There is not enough evidence to suggest that the mean life span today is greater than {population_mean_hypothesized} years.")
```

### Visualize the t-distribution and the critical region

```python
t_values = np.linspace(-4, 4, 1000)
pdf_values = stats.t.pdf(t_values, df)
plt.plot(t_values, pdf_values, label='t-distribution')

# Highlight the critical region
plt.fill_between(t_values, pdf_values, where=(t_values > t_critical), color='red', alpha=0.5, label='Critical Region')

# Highlight the observed t-statistic
plt.axvline(x=t_statistic, color='blue', linestyle='--', label='Observed t-statistic')

plt.title('Hypothesis Test and Critical Region')
plt.xlabel('t-values')
plt.ylabel('Probability Density Function')
plt.legend()
plt.show()
```

### Method 2 By Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given data
sample_mean = 71.8
population_std_dev = 8.9
sample_size = 100
population_mean_hypothesized = 70
significance_level = 0.05

# Calculate the test statistic and p-value
t_statistic, p_value = stats.ttest_1samp(np.random.normal(population_mean_hypothesized, population_std_dev, sample_size), sample_mean)

# Degrees of freedom
df = sample_size - 1

# Critical t-value for a one-tailed test
t_critical = stats.t.ppf(1 - significance_level, df)

# Generate t-values for the plot
t_values = np.linspace(-4, 4, 1000)
pdf_values = stats.t.pdf(t_values, df)

# Plot the t-distribution
plt.plot(t_values, pdf_values, label='t-distribution')

# Highlight the critical region
plt.fill_between(t_values, pdf_values, where=(t_values > t_critical), color='red', alpha=0.5, label='Critical Region')

# Highlight the observed t-statistic
plt.axvline(x=t_statistic, color='blue', linestyle='--', label='Observed t-statistic')

# Add labels and legend
plt.title('Hypothesis Test and Critical Region')
plt.xlabel('t-values')
plt.ylabel('Probability Density Function')
plt.legend()

# Show the plot
plt.show()

# Print the test information
print("Test Information:")
print(f"Sample Mean: {sample_mean}")
print(f"Population Standard Deviation: {population_std_dev}")
print(f"Sample Size: {sample_size}")
print(f"Population Mean (Hypothesized): {population_mean_hypothesized}")
print(f"Significance Level: {significance_level}")
print(f"Degrees of Freedom: {df}")
print(f"Critical t-value: {t_critical}")
print(f"Observed t-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Decision based on the p-value
if p_value < significance_level:
    print(f"\nReject the null hypothesis. There is evidence to suggest that the mean life span today is greater than {population_mean_hypothesized} years.")
else:
    print(f"\nFail to reject the null hypothesis. There is not enough evidence to suggest that the mean life span today is greater than {population_mean_hypothesized} years.")
```

**Example**
A manufacturer of sports equipment has
developed a new synthetic fishing line that the
company claims has a mean breaking strength of 8
kilograms with a standard deviation of 0 5 kilogram
Test the hypothesis that μ 8 kilograms against the
alternative that μμ≠8 kilograms if a random sample of
50 lines is tested and found to have a mean breaking
strength of 7 8 kilograms Use a 0 01 level of
significance

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given data
sample_mean = 7.8
population_std_dev = 0.5
sample_size = 50
population_mean_hypothesized = 8
significance_level = 0.01

# Hypotheses
# Null hypothesis (H0): The mean breaking strength is equal to 8 kilograms.
# Alternative hypothesis (H1): The mean breaking strength is not equal to 8 kilograms.

# Calculate the test statistic
z_statistic = (sample_mean - population_mean_hypothesized) / (population_std_dev / np.sqrt(sample_size))

# Degrees of freedom
df = sample_size - 1

# Critical z-values for a two-tailed test
z_critical_lower = stats.norm.ppf(significance_level / 2)
z_critical_upper = -z_critical_lower

# Print the hypotheses
print(f"Null Hypothesis (H0): μ = {population_mean_hypothesized} (The mean breaking strength is equal to {population_mean_hypothesized} kilograms.)")
print(f"Alternative Hypothesis (H1): μ ≠ {population_mean_hypothesized} (The mean breaking strength is not equal to {population_mean_hypothesized} kilograms.)\n")

# Print the results and decision
print(f"Sample Mean: {sample_mean}")
print(f"Population Standard Deviation: {population_std_dev}")
print(f"Sample Size: {sample_size}")
print(f"Population Mean (Hypothesized): {population_mean_hypothesized}")
print(f"Significance Level: {significance_level}")
print(f"Degrees of Freedom: {df}")
print(f"Critical z-values: {z_critical_lower} and {z_critical_upper}")
print(f"Observed z-statistic: {z_statistic}")

# Decision based on the comparison
if np.abs(z_statistic) > max(np.abs(z_critical_lower), np.abs(z_critical_upper)):
    print(f"\nReject the null hypothesis. There is evidence to suggest that the mean breaking strength is not equal to {population_mean_hypothesized} kilograms.")
else:
    print(f"\nFail to reject the null hypothesis. There is not enough evidence to suggest that the mean breaking strength is different from {population_mean_hypothesized} kilograms.")
```

```python
# Visualize the normal distribution and the critical region
z_values = np.linspace(-4, 4, 1000)
pdf_values = stats.norm.pdf(z_values)
plt.plot(z_values, pdf_values, label='Normal Distribution')

# Highlight the critical region
plt.fill_between(z_values, pdf_values, where=((z_values < z_critical_lower) | (z_values > z_critical_upper)), color='red', alpha=0.5, label='Critical Region')

# Highlight the observed z-statistic
plt.axvline(x=z_statistic, color='blue', linestyle='--', label='Observed z-statistic')

plt.title('Hypothesis Test and Critical Region')
plt.xlabel('z-values')
plt.ylabel('Probability Density Function')
plt.legend()
plt.show()
```
